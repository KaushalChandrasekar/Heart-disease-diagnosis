Load and Display the Dataset

```python
import pandas as pd
# Check TensorFlow version
import tensorflow as tf
print(tf.__version__)
df = pd.read_csv('/content/Heart Disease data.csv')
print(df.head())
print(df.describe())
```

Visualize distribution of Key features

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Distribution of Target Variable

```python
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df) 
plt.title('Distribution of Heart Disease Presence')
plt.show()
```

2. Correlations

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

3. Heart Disease by Gender and Age

```python
plt.figure(figsize=(12,5))
sns.countplot(x='age', hue='sex', data=df[df['target']==1]) 
plt.title('Heart Disease Presence by Gender and Age')
plt.xticks(rotation=45)
plt.show()
```

4. Chest Pain Type Analysis

```python
plt.figure(figsize=(8,6))
sns.countplot(x='cp', data=df[df['target']==1]) 
plt.title('Heart Disease Presence by Chest Pain Type')
plt.show()
```

5. Risk Factors Analysis
For Cholesterol Levels

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='target', y='chol', data=df) 
plt.title('Heart Disease Presence by Cholesterol Levels')
plt.show()
```

## The visualizations generated from the heart disease dataset offer crucial insights:

Distribution of Heart Disease Presence: The almost equal distribution suggests a balanced dataset, which is beneficial for any predictive modeling.

Correlation Matrix: There are several notable correlations, such as maximum heart rate (thalach) with the target, and ST depression (oldpeak) with the slope of the peak exercise ST segment (slope). These correlations could be indicators of higher risk for heart disease.

Heart Disease Presence by Gender and Age: This chart illustrates that heart disease occurrence varies across age and gender. It appears that in certain age groups, one gender is more affected than the other.

Heart Disease Presence by Chest Pain Type: The number of people with heart disease varies by chest pain type, with some types being more common among affected individuals.

Heart Disease Presence by Cholesterol Levels: Cholesterol levels might not show a distinct separation between individuals with and without heart disease, but outliers could indicate that extremely high or low levels contribute to the risk.

6. Pairwise Relationships

```python
sns.pairplot(df, hue='target', vars=['age', 'trestbps', 'chol', 'thalach', 'oldpeak'])
plt.suptitle('Pairwise Relationships', size=16)
plt.show()
```

7. Age Group Analysis

```python
age_bins = [20, 30, 40, 50, 60, 70, 80]
df['age_group'] = pd.cut(df['age'], bins=age_bins)
plt.figure(figsize=(10,6))
sns.countplot(x='age_group', hue='target', data=df)
plt.title('Heart Disease Presence by Age Group')
plt.show()
```

8. Exercise Related Features

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='oldpeak', y='thalach', hue='target', data=df)
plt.title('ST Depression vs. Max Heart Rate by Heart Disease Presence')
plt.show()
```

9. Thalassemia Analysis

```python
plt.figure(figsize=(8,6))
sns.countplot(x='thal', hue='target', data=df)
plt.title('Heart Disease Presence by Thalassemia Type')
plt.show()
```

10. Major Vessels Analysis

```python
plt.figure(figsize=(8,6))
sns.countplot(x='ca', hue='target', data=df)
plt.title('Heart Disease Presence by Number of Major Vessels')
plt.show()
```

## The additional visualizations reveal further insights into the heart disease dataset:

Pairwise Relationships: This detailed view highlights how each pair of features correlates and potentially interacts with the target variable, offering a multi-dimensional perspective on the data.

Heart Disease Presence by Age Group: The breakdown by age group illustrates a trend where the presence of heart disease peaks in certain age groups before declining, which could reflect various life stage-related risk factors.

ST Depression vs. Max Heart Rate by Heart Disease Presence: This scatter plot suggests a potential pattern where higher maximum heart rates are associated with lower ST depression values in the context of heart disease.

Heart Disease Presence by Thalassemia Type: The countplot shows a higher prevalence of heart disease in individuals with a reversable defect (thal value of 2), indicating that this could be a significant risk factor.

Heart Disease Presence by Number of Major Vessels: The presence of heart disease seems to decrease as the number of major vessels observed increases, which might indicate that fewer major vessels observed during fluoroscopy could be a warning sign.
